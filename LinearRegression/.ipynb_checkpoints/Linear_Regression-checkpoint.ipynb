### Implementation of the linear model
The linear model implementation includes the functionalities of normalizing the data by column mean and standard deviation. It also has the options of using gradient descendent or normal equation from ordinary least square to calculate the regression coefficients. 
The gradient descendent optimization was accomplished by BFGS method implemented in scipy.optimization package.

```python
class LinearReg(object):
    def __init__(self,normalize=True, mode='gradient'):
        '''
        input:
           normalized: if the data need to be normalized
           mode: using gradient descent or normal equation to find theta
        '''
        self.normalize=normalize
        self.mode=mode
        
    def _normalize(self,X):
        '''
        input:
          X: feature matrix for training dataset
          
        output:
          returns the normalized training dataset, and set up the column means and std for
          normalizing test dataset
        '''
        self.feature_center=X.mean(axis=0)
        self.feature_std=X.std(axis=0)
        return (X-self.feature_center)/self.feature_std
        
    def _costCompute(self,theta,X,y):
        '''
        input: 
           X: n times m matrix, n is the number of observations, and m is the number of features, including intecept
           y: n times 1 vector
           theta: m times 1 vector, corresponding to the regression coefficients, including intercept
       
        output:
           a scalar variable corresponding to the cost
        '''    
        num_obs=float(len(y))        
        return 0.5*(1/num_obs)*((y-X.dot(theta))**2).sum()

    def _gradCompute(self,theta,X,y):
        '''
        input:
           X: n times m matrix, n is the number of observations, and m is the number of features, including intercept
           y: n times 1 vector
           theta: m times 1 vector, corresponding to the regression coefficients of m features, including intercept
           
        output:
           returns the gradient of the theta parameter
        '''
        num_obs=len(y)   
        return -1.0/num_obs*X.T.dot(y-X.dot(theta)) 
    
    def fit(self,X,y):
        if self.mode=='normal':
            self._fit_by_normal(X,y)
        else:            
            self._fit_by_gradient(X,y)


    def _fit_by_gradient(self,X,y):
        '''
        input:
           X: n times m matrix, n is the number of observations, and m is the number of features, including intercept
           y: n times 1 target variable vector
           
        output:
          fit the data by linear regression using gradient desendent, and set the _theta values
        '''
        num_obs=X.shape[0]
        num_features=X.shape[1]
        
        #normalize the training dataset
        if self.normalize:
            X=self._normalize(X)
            
        # insert intercept column
        X=np.hstack((np.ones((num_obs,1)),X))
        
        #set the initial values for theta
        theta=np.zeros(num_features+1)
        
        #find the optimum theta values that minimize cost function
        rs=optimize.minimize(fun=self._costCompute, x0=theta, jac=self._gradCompute, args=(X,y),method='BFGS')
        print(rs.message)
        if rs.success:
            self._theta=rs.x
            
    def _fit_by_normal(self,X,y):
        num_obs=X.shape[0]
        if self.normalize:
            X=self._normalize(X)
        X=np.hstack((np.ones((num_obs,1)),X))    
        self._theta=np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
           
    def predict(self,X_new):
        '''
        input: 
          X_new a n time m feature matrix for a new dataset to be predicted by model
          
        output:
          predicted target variable vector (n time 1)
        '''
        num_obs=X_new.shape[0]
        
        if self.normalize:
            X_new=(X_new - self.feature_center)/self.feature_std
                           
        return np.hstack((np.ones((num_obs,1)),X_new)).dot(self._theta)    
    
```

### Testing the linear model

#### Test case 1: simple linear regression (one predictor)

```python
X=profit['population'].values.reshape((97,1))
y=profit['profit'].values
new_data=np.array([[3.5]])
```

First, we use the gradient descendent optimization to find the regression coefficients, and use the fitted model to predict the profit of a new case

```python
lrg1=LinearReg()
lrg1.fit(X,y)
lrg1.predict(new_data)
```

Next, we test the model by caculating the regression coefficients using ordinary least square equation, and used the fitted model to predict the profit of a new case

```python
lrn1=LinearReg(mode='normal')
lrn1.fit(X,y)
lrn1.predict(new_data)
```

Conclusion: By comparing the results obtained by gradient descendent and ordinary least square equation, we can see that the results are consistent.

#### Test case 3: Boston dataset containing multiple predictors

First, we load the Boston dataset from sklearn and do some simple data exploaration to find the relationship among the predictors, if there is any missing values, and explore features that might be interesting.

```python
from sklearn.datasets import load_boston
boston=load_boston()
```

```python
X_boston=pd.DataFrame(boston.data,columns=boston.feature_names)
y_boston=boston.target
```

```python
X_boston.head()
```

```python
X_boston.info()
```

```python
sns.pairplot(data=X_boston)
```

Pairplot showed many predictor variables are correlated. Representative features may be selected or extracted for better modeling performacne. In addition, some numerical predictors are not normally distributed, which may be transformed to normal distributions for better modeling performance. In this project, data transformation, feature selection and extraction, were not performed, since the purpose here to test the linear regression model implemented in this project.

Next, we load the train_test_split and metrics from sklearn to split the dataset and evaluate the prediction results.

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

We then split the dataset to training and test datasets, and then fit the training dataset using gradient descendent optimization, and predict the test dataset

```python
X_train,X_test,y_train,y_test=train_test_split(X_boston,y_boston,test_size=0.3,random_state=101)

lr3=LinearReg()
lr3.fit(X_train,y_train)

y_pred_g=lr3.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred_g))
```

Now, we can plot the predicted versus observed y values

```python
plt.scatter(y_test,y_pred_g)
plt.xlabel("observed values")
plt.ylabel("predicted values")
```

Then we test the model using the ordinary least square equation to fit the training dataset, predict the test dataset, and plot the predicted values versus the observed values

```python
lr4=LinearReg(mode='normal')
lr4.fit(X_train,y_train)
y_pred=lr4.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("observed values")
plt.ylabel("predicted values")
```

We obtained the mean standard deviation of the error, to evalute the significance of this error, we can compare this value to the mean of the target variable

```python
y_test.mean()
```

Comparing the mean value of the target variable and the standard deviation of the error, the error is about 20% of the mean target variable.

### Comparison to sklearn implementation
Finally, we compare the results obtained by the linear regression model implemented by sklearn to the results obtained by the model implemented in this project

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
pl=Pipeline([('scale',StandardScaler()),('lr',LinearRegression())])
pl.fit(X_train,y_train)
y_pred_sk=pl.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred_sk))
```

```python
plt.scatter(y_test,y_pred_sk)
plt.xlabel("observed values")
plt.ylabel("predicted values")
```

Conclusion: Comparing the model implemented by gradient descent optimization, ordinary least square equation and the implementation of sklearn package, consistent results were obtained in terms of the mean squared standard deviations
